```
import numpy as np

a1 = np.array([0,1,1,1])
a2 = np.array([1,0,1,1])
a3 = np.array([2,0,2,2])
```

```
def cosine_sim(a,b):
    dot = np.matmul(a,b)
    nom1,nom2 = np.linalg.norm(a),np.linalg.norm(b)
    return dot/(nom1*nom2)
```

```
sim1 = cosine_sim(a1,a2)
print(round(sim1,3))

sim2 = cosine_sim(a1,a3)
print(round(sim2,3))

sim3 = cosine_sim(a2,a3)
print(round(sim3,3))
sims = [a1,a2,a3]
```
